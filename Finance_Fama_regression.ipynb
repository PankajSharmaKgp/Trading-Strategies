```python
import pandas_datareader.data as reader
import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
end=dt.date(2020,6,30)
start=dt.date(end.year-5,end.month,end.day)
funds=['FDGRX']
```

```python
fundret=reader.get_data_yahoo(funds,start,end)['Adj Close'].pct_change()
```

```python
fundmon=fundret.resample('M').agg(lambda x: (x+1).prod()-1)
```

```python
fundmon
```

```python
factors=reader.DataReader('F-F_Research_Data_Factors','famafrench',start,end)[0]
```

```python
factors
```

```python
#make the indices of both dataframes same to merge
fundmon.index=factors.index
```

```python
fundmon
```

```python
import pandas as pd
merge=pd.merge(fundmon,factors,on='Date')
```

```python
merge
```

```python
merge[['Mkt-RF','SMB','HML','RF']]=merge[['Mkt-RF','SMB','HML','RF']]/100
```

```python
merge
```

```python
merge['FDGRX-RF']=merge.FDGRX-merge.RF
```

```python
merge
```

```python
y=merge['FDGRX-RF']
X=merge[['Mkt-RF','SMB','HML']]

X_sm=sm.add_constant(X)
```

```python
model=sm.OLS(y,X_sm)
```

```python
results=model.fit()
```

```python
results.summary()
```
