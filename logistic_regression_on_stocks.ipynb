```python
import yfinance as yf
import pandas as pd
import statsmodels.api as sm
```

```python
df=yf.download('^GSPC',start='2001-01-03',end='2005-12-31')
```

```python
df
```

```python
df1=df['Adj Close'].pct_change()*100
```

```python
df1
```

```python
df1=df1.rename('Today')
```

```python
df1
```

```python
df1=df1.reset_index()
```

```python
df1
```

```python
#shifting ..that is getting the prices for lag 1, lag 2 for 1 day two day and so on before today
for i in range(1,6):
    df1['Lag' + str(i)]=df1['Today'].shift(i)
```

```python
#getting one day prior volumes
df1['Volume']=df.Volume.shift(i).values/1000000000
```

```python
df1
```

```python
df1=df1.dropna()
```

```python
df1
```

```python
#take 1 for up move and 0 for downmove
df1['Direction']=[1 if i>0 else 0 for i in df1['Today']]
```

```python
df1
```

```python
df1=sm.add_constant(df1)
```

```python
df1
```

```python
X=df1[['const','Lag1','Lag2','Lag3','Lag4','Lag5','Volume']]
y=df1['Direction']      
```

```python
model=sm.Logit(y,X)
```

```python
results=model.fit()
```

```python
results.summary()
```

```python
results.predict(X)
```
