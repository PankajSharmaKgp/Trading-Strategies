```python
import pandas as pd
import datetime as dt
import yfinance as yf
from pandas.tseries.offsets import MonthEnd
```

```python
tickers=pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2]
```

```python
tickers
```

```python
tickers=tickers.Symbol.to_list()
```

```python
tickers
```

```python
df=yf.download(tickers,start='2009-01-01')
```

```python
df
```

```python
price=df['Adj Close']
```

```python
#convert the string datetime index to numeral datetime index
price.index=pd.to_datetime(price.index)
```

```python
mtl_ret=price.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)
```

```python
mtl_ret
```

```python
#next we choose a random date and go one month before and check the worst performing stocks and buy em
formation=dt.datetime(2009,2,28)
ret_1=mtl_ret.loc[formation-MonthEnd(1)]
```

```python
ret_1
```

```python
ret_1.name='mtl_ret'
```

```python
ret_1=pd.DataFrame(ret_1)
```

```python
ret_1
```

```python
ret_1['decile']=pd.qcut(ret_1.mtl_ret,10,labels=False)
```

```python
ret_1
```

```python
loosers=ret_1[ret_1.decile==0].index
```

```python
loosers
```

```python
los_ret=mtl_ret.loc[formation,mtl_ret.columns.isin(loosers)].mean()
```

```python
los_ret
```

```python
#backtest the strategy for all months
def reversal(formation):
    ret_1=mtl_ret.loc[formation-MonthEnd(1)]
    ret_1.name='mtl_ret'
    ret_1=pd.DataFrame(ret_1)
    ret_1['decile']=pd.qcut(ret_1.mtl_ret,10,labels=False,duplicates='drop')
    loosers=ret_1[ret_1.decile==0].index
    los_ret=mtl_ret.loc[formation,mtl_ret.columns.isin(loosers)].mean()
    return los_ret
    
```

```python
#get returns for all dates starting from third date
returns=[]
dates=[]
#i is the formation date
for i in mtl_ret.index[2:]:
    returns.append(reversal(i))
    dates.append(i)
```

```python
returns
```

```python
frame=pd.DataFrame({'dates':dates,'returns':returns})
```

```python
frame
```

```python
frame.returns.mean()
```
