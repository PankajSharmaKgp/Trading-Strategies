```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as reader
from dateutil.relativedelta import relativedelta
```

```python
stickers=['AAPL',
'AMGN',
'AXP',
'BA',
'CAT',
'CRM',
'CSCO',
'CVX'
'DIS',
'DOW',
'GS',
'HD',
'HON',
'IBM',
'INTC',
'JNJ',
'JPM',
'KO',
'MCD',
'MMM',
'MRK',
'MSFT',
'NKE',
'PG',
'TRV',
'UNH',
'V',
'VZ',
'WBA',
'WMT']
```

```python
start=dt.datetime(2018,1,31)
end=dt.datetime(2020,1,31)
returns=reader.get_data_yahoo(stickers,start,end)['Adj Close']
```

```python
returns
```

```python
mtl_ret=returns.pct_change().resample('M').agg(lambda x:(x+1).prod()-1)
```

```python
mtl_ret
```

```python
import numpy as np
past_11=(mtl_ret+1).rolling(11).apply(np.prod)-1
```

```python
past_11
```

```python
formation=dt.datetime(2019,12,31)
end_measurement=formation-relativedelta(months=1)
```

```python
end_measurement
```

```python
ret_12=past_11.loc[end_measurement]
```

```python
ret_12=ret_12.reset_index()
```

```python
ret_12
```

```python
ret_12['quintile']=pd.qcut(ret_12.iloc[:,1],5,labels=False)
```

```python
ret_12
```

```python
winners=ret_12[ret_12['quintile']==4]
loosers=ret_12[ret_12['quintile']==0]
```

```python
winners
```

```python
winnerret=mtl_ret.loc[formation+relativedelta(months=1),mtl_ret.columns.isin(winners.Symbols)]
```

```python
looserret=mtl_ret.loc[formation+relativedelta(months=1),mtl_ret.columns.isin(loosers.Symbols)]
```

```python
winnerret
```

```python
looserret
```

```python
momentumret=winnerret.mean()-looserret.mean()
```

```python
momentumret
```
